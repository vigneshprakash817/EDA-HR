```python
plt.plot(df['Age'],df['Tenure_Years'])
plt.title('Age wise tenure')
plt.xlabel('Age')
plt.ylabel('Tenure_Years')
plt.grid(True)
plt.show()
```

```python
plt.bar(df['Department'],df['Salary'])
plt.title('Department wise Salary')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
plt.hist(df['Salary'],bins=3)
plt.title('Salary Bin')
plt.xlabel('Bin')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
plt.pie(df['Salary'],labels=df['Gender'])
plt.title('Gender wise salary',fontdict={'fontsize':20,'fontname':'arial','fontweight':'bold','color':'blue'})
plt.xlabel('Salary')
plt.ylabel('Gender')
plt.grid(True)
plt.show()
plt.figure(figsize=(8,6))
```
